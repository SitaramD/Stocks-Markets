```python
#!pip install numpy
#!pip install pandas
#!pip install datetime
#!pip install matplotlib.plt
#!pip install mplfinance as mpf
#!pip install seaborn as sns
#!pip install talib as ta
#!pip install math
#!pip install yfinance
# import libraries
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import mplfinance as mpf
import seaborn as sns
import talib
import datetime
import yfinance as yf
df1= pd.read_csv("D:\\Python\\Pycharm\\NIFTYSMALL50CAP.csv")
#ticker_list = df1["Symbol"].tolist()
for i in range(len(df1)):
    ticker = df1["Symbol"][i]
    # Downloading BAnk Nifty data from Yahoo Financeh
    df= yf.download(ticker , period ="2d", interval="5m")
    df.to_csv("data.csv")
    df = pd.read_csv("data.csv")
    for i in range(len(df)):
        print("Stock Name =", ticker)
        # Considering only closing values for predicting trend
        Close = df["Close"]
        # Filling NAN vales backwards for Dataframe
        df =df.fillna( method = "bfill")
        df.rename(columns = {'Unnamed: 0':'Datetime'}, inplace = True)
        # Generating ATR Values for 7 days
        df["Atr"] = talib.ATR(df["High"].values,df["Low"].values,df["Close"].values, timeperiod = 7)
        # Calculating  RSI values
        period =7
        source =df["Close"]
        df["RSI"] = talib.RSI(source, timeperiod=period)
        # Filling NAN vales in Diff column backwards for Dataframe
        df =df.fillna(method = "bfill")
        # defining dataframe as nan values
        df["Buy Entry Price"] = np.nan # Storing in a datframe
        df["Sell Entry Price"] = np.nan #[0] # Storing in a datframe
        TotalBuypositions = 0
        TotalSellpositions = 0
        #position = int
        for i in range(len(df)-1):
            if  df["RSI"][i+1] > 70 and  df["RSI"][i] > 60:
                print("**********************")

            elif df["RSI"][i] < 30 and df["RSI"][i+1] >20:
                print("Stock Name = ", ticker)
                print("Buy  @", ticker,df["Close"][i])
                df["Sell Entry Price"][i] = df["Close"][i]  # Storing in a datframe
                df["Buy Entry Price"][i] = 0.00             # Storing in a datframe
                print("Entry Time:  ",df["Datetime"][i])
                print("Target", df["Close"][i] + 1.5 * df["Atr"][i]) # Risk Reward Ratio 1:2, Take profit = 4 X ATR for 5 minute timeframe
                print("Stoploss", df["Close"][i] - df["Atr"][i]) # Risk Reward Ratio 1:2, Stop loss = 2 X ATR for 5 minute timeframe
                TotalSellpositions = TotalSellpositions + 1
                print("TotalBuypositions =", TotalSellpositions)
                print("**********************")
            else:
                print("******")



        else:
            print("No Call on :", df["Datetime"][i])

else:
    print("Total Trades =", TotalSellpositions + TotalBuypositions)
    print(" Next Stock")

```
