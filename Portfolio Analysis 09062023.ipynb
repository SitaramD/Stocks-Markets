```python
import yfinance as yf
import pandas as pd

# Assuming you have a list of tickers and a dictionary of start dates for each stock named start_dates
tickers = ["AAPL", "KMB", "SNCE", "HMPT", "BLBD", "PTGX", "DZSI",
           "TRTN", "HLT", "WSC", "ARL", "PRAX", "RETA", "BIIB", "WEJO", "ATIP", "VRAY"]

# Define the start dates for each stock
start_dates = {
    'AAPL': '2023-04-21',
    'KMB': '2023-04-21',
    'SNCE': '2023-04-21',
    'HMPT': '2023-04-21',
    'BLBD': '2023-04-21',
    'PTGX': '2023-04-21',
    'DZSI': '2023-04-21',
    'TRTN': '2023-04-21',
    'HLT': '2023-04-21',
    'WSC': '2023-04-21',
    'ARL': '2023-04-21',
    'PRAX': '2023-04-21',
    'RETA': '2023-04-21',
    'BIIB': '2023-04-21',
    'WEJO': '2023-04-21',
    'ATIP': '2023-04-21',
    'VRAY': '2023-04-21'
}

# Define the quantities for each stock
quantities = {
    'AAPL': 100,
    'KMB': 50,
    'SNCE': 75,
    'HMPT': 30,
    'BLBD': 60,
    'PTGX': 90,
    'DZSI': 40,
    'TRTN': 80,
    'HLT': 70,
    'WSC': 55,
    'ARL': 25,
    'PRAX': 45,
    'RETA': 85,
    'BIIB': 65,
    'WEJO': 35,
    'ATIP': 20,
    'VRAY': 50
}

# Combine the dictionaries of stocks and quantities
stocks_quantities = dict(zip(tickers, quantities.values()))

# Initialize empty DataFrames for storing the results
cumulative_returns = pd.DataFrame()
drawdown = pd.DataFrame()
max_drawdown = pd.Series()

# Fetch daily closing prices for the stocks
data = yf.download(list(stocks_quantities.keys()), start="2023-04-21")['Close']

# Perform portfolio analysis for each stock
for stock, start_date in start_dates.items():
    # Filter the DataFrame based on the start date for the current stock
    df_stock = data[stock][data.index >= start_date]

    # Calculate daily returns for the current stock
    returns = df_stock.pct_change()

    # Calculate cumulative returns for the current stock
    cumulative_returns[stock] = (returns + 1).cumprod()

    # Calculate drawdown for the current stock
    rolling_max = cumulative_returns[stock].expanding().max()
    drawdown[stock] = (cumulative_returns[stock] / rolling_max) - 1

    # Calculate maximum drawdown for the current stock
    max_drawdown[stock] = drawdown[stock].min()

# Calculate the portfolio value for each date
portfolio_value = (data * pd.Series(stocks_quantities)).sum(axis=1)

# Calculate the portfolio returns
portfolio_returns = portfolio_value.pct_change()

# Calculate the cumulative portfolio returns
cumulative_portfolio_returns = (portfolio_returns + 1).cumprod()

# Print the results
print("Cumulative Returns:")
print(cumulative_returns.tail())
print("\nDrawdown:")
print(drawdown.tail())
print("\nMaximum Drawdown:")
print(max_drawdown)
print("\nPortfolio Value:")
print(portfolio_value.tail())
print("\nPortfolio Returns:")
print(portfolio_returns.tail())
print("\nCumulative Portfolio Returns:")
print(cumulative_portfolio_returns.tail())
```
