```python
#!pip install numpy 
#!pip install pandas
#!pip install datetime
#!pip install matplotlib.plt
#!pip install mplfinance as mpf
#!pip install seaborn as sns
#!pip install talib as ta
#!pip install math
#!pip install yfinance
# import libraries
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import mplfinance as mpf
import seaborn as sns
import talib
import datetime
import yfinance as yf
df1= pd.read_csv("NIFTYINDEX.csv")
ticker_list = df1["Symbol"].tolist()
for i in range(len(ticker_list)):
        ticker = df1["Symbol"][i]
        # Downloading Bank Nifty data from Yahoo Finance
        df= yf.download(ticker , period = "2021-06-23", interval="1m")
        df.to_csv("data.csv")
        df = pd.read_csv("data.csv")
        print("Stock Name =", df1["Symbol"][i])
        # Considering only closing values for predicting trend
        Close = df["Close"]
        # Extracting 3 sma using talib library
        sma3 = talib.SMA(Close, timeperiod = 3)
        # Extracting 8 sma using talib library
        sma8 = talib.SMA(Close, timeperiod = 8)
        # Creating a new column by incorporating the difference between SMA3 and SMA8
        df["sma3"] = sma3
        df["sma8"] = sma8
        # Filling NAN vales backwards for Dataframe
        df =df.fillna( method = "bfill")
        df.rename(columns = {'Unnamed: 0':'Datetime'}, inplace = True)
        df["Diff"] = sma3-sma8
        # Generating ATR Values for 3 days
        df["Atr"] = talib.ATR(df["High"].values,df["Low"].values,df["Close"].values, timeperiod = 3)
        # Initialize Bollinger Bands Indicator
        period = 20 # We can optimise it
        stdNbr = 2 # We can optimise it
        # Moving Averages , simple
        indicator_bb = talib.BBANDS(df["Close"].values, timeperiod = period, nbdevup=stdNbr,nbdevdn=stdNbr,matype=0)                            
        df["Upperband"] = indicator_bb[0]
        df["Middleband"] = indicator_bb[1]
        df["Lowerband"] = indicator_bb[2]
        # MACD Indicator
        macd, signal, hist = talib.MACD(df['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
        # Converting pandas to dataframe and adding to df
        df["MACD"] = macd
        df["MACDSignal"] = signal
        df["MACDHist"] = hist
        # On Balnce Volume Indicator
        obv= talib.OBV(df["Close"],df["Volume"])
        df["OBV"] = obv
        #MFI' in indicator:
        period = 14 # We can optimize it
        mfi = talib.MFI(df["High"], df["Low"], df["Close"], df["Volume"], timeperiod= period)
        # Adding column to df
        df["MFI"] = mfi
        # Calculating  RSI values 
        period =6
        source =df["Close"]
        df["RSI"] = talib.RSI(source, timeperiod=period)
        # William % Indicator
        williamvalue = talib.WILLR(df["High"], df["Low"], df["Close"])
        df["William%"] = williamvalue
        # Filling NAN vales in Diff column backwards for Dataframe
        df =df.fillna( method = "bfill")
        # Using a for loop to compare 3sma with 8sma and generating a signal.
        df["Buy Entry Price"] = np.nan # Storing in a datframe
        df["Sell Entry Price"] = np.nan #[0] # Storing in a datframe 
        #position = int
        #i = len(df)-10
        df=df.round(2)
        TotalSellpositions = 0
        for i in range(len(df)):
            if  df["Atr"][i] > df["Atr"][i+1] and df["RSI"][i] > df["RSI"][i+1]:
                print("INDEX is in  Bearish Trend :", ticker)
                print(" Sell Call @ ", df["Close"][i])
                print(" Entry Time:  ", df["Datetime"][i]) 
                print(" Take Profit= ", df["Close"][i] - 1.2*df["Atr"][i])
                print(" Stop Losss @ ", df["Close"][i] + 0.60*df["Atr"][i])
                TotalSellpositions = TotalSellpositions + 1
                print(" Total Sell Positions= ", TotalSellpositions)
                                   
                print("**********************") 
            #elif df["Atr"][i] < df["Atr"][i+1] or df["RSI"][i] < df["RSI"][i+1]:
                #print("Buy Call")
                #print("Entry Time:  ", df["Datetime"][i])                     
                #print("**********************")                
            #else:
                #print("HOLD Call")
                #print("Entry Time:  ", df["Datetime"][i])                     
                #print("**********************")
else:
    print(" End of Excecution")
    print(" Total Sell Positions= ", TotalSellpositions)
```
