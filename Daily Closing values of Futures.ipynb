```python
# Importing Libraries
import numpy as np
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# The Futures we'll use for correlation
df= pd.read_csv("D:\\Venket\\HIST PRICES\\January\\01012023\\quotes.csv")
```

```python
df
```

```python
# Converting dataframe to List
Fut_list =df["Symbol"].values.tolist()
```

```python
# Printing List
Fut_list
```

```python
# Calculation to grab all the closing prices for the tech stock list into one DataFrame
closing_df = yf.download(Fut_list, period = "1mo", interval ="1d")['Adj Close']
```

```python
# Check last 4 days
closing_df.tail(4)
```

```python
# Finding Correlation
closing_df.corr().round(2)
```

```python
matrix= closing_df.corr().round(2)
matrix=matrix.unstack()
k=matrix[matrix=0.00]
```

```python
k=matrix[matrix==-0.00]
```

```python
k
```

```python
df1 = pd.DataFrame(matrix) 
    
df1.to_csv('0 Uncorrelated Data.csv') 
```

```python
# Generating Heatmap
sns.heatmap(closing_df.corr().round(2), vmin=-1, vmax=1, annot=True)
```
