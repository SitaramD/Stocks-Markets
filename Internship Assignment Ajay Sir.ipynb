```python
# Defining a function
def fetch_detail(stockcode):
    
        #print(stockcode)
        stock_url = 'https://www.nseindia.com/live_market/dynaContent/live_watch/get_quote/GetQuote.jsp?symbol='+str(stockcode)
        print(stock_url)
        headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'}
        response = requests.get(stock_url, headers=headers)
        response
        soup = BeautifulSoup(response.text, 'html.parser')
        data_array = soup.find(id='responseDiv').getText().strip().split(":")
        #type (data_array)
        #print(data_array)
        for item in data_array:
             if 'symbol' in item:
                index = data_array.index(item)+1
                print("Index -> "+ str(index))
                symbol1=data_array[index].split('"')[1]
                symbol.append(symbol1)
                #print("symbol = ",symbol)
        # Extracting last traded Price
        for item in data_array:
            if 'lastPrice' in item:
                index = data_array.index(item)+1
                print("Index -> "+ str(index))
                lastPrice1=data_array[index].split('"')[1]
                lastPrice.append(lastPrice1)
                #print("latest Price = ",latestPrice)
                        # Extracting Total buy Quantity
        for item in data_array:
            if 'totalBuyQuantity' in item:
                index = data_array.index(item)+1
                print("Index -> "+ str(index))
                totalBuyQuantity1=data_array[index].split('"')[1]
                totalBuyQuantity.append(totalBuyQuantity1)
                #print("totalBuyQuantity = ",totalBuyQuantity)
                # Extracting Total sell Quantity
        for item in data_array:
             if 'totalSellQuantity' in item:
                index = data_array.index(item)+1
                print("Index -> "+ str(index))
                totalSellQuantity1=data_array[index].split('"')[1]
                totalSellQuantity.append(totalSellQuantity1)
                #print("totalSellQuantity = ",totalSellQuantity)
        Values =(symbol, lastPrice, totalBuyQuantity, totalSellQuantity)        
        return Values

    #print(datetime.now())

```
