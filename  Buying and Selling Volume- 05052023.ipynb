```python
import yfinance as yf
import pandas as pd
from datetime import datetime

# Define a list of stock symbols
stocks = tickers

# Get today's date
today = datetime.now().strftime("%Y-%m-%d")

# Create an empty DataFrame to store the results
results = pd.DataFrame(columns=['Symbol', 'Date', 'Close','Volume'])

# Loop through the list of stocks
for stock in stocks:
    # Download the stock data for the past 6 months
    data = yf.download(stock, period="6mo", interval="1d")

    # Calculate the 30-day average volume
    data['avg_volume'] = data['Volume'].rolling(window=30).mean()

    # Identify the days where the buying volume is twice the average 30-day volume
    data['buying_volume'] = (data['Close'] > data['Open']) * data['Volume']
    data['buying_volume_2x_avg'] = (data['buying_volume'] > (data['avg_volume']*1.25))

    # Check if today's buying volume is high
    if data['buying_volume_2x_avg'].iloc[-1]:
        # Append the results to the dataframe
        results = pd.concat([results, pd.DataFrame({'Symbol': stock, 'Date': today, 'Close': data['Close'], 'Volume': data['Volume'].iloc[-1]}, index=[0])], ignore_index=True)
```

```python
# Print the results: Stocks having daily volume > 1.25 times 30 Day rollimg Average Volume.
print(results)
```
