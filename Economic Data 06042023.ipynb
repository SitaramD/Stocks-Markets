```python
from fredapi import Fred
fred = Fred(api_key='7086f29199d2614176264979fb0bebb9')
```

```python
#Starting gand  ending duration of data
start="2022-04-06"
end="2023-04-07"
```

```python
list1= df3.values.tolist()
```

```python
print(list1)
```

```python
# To pull values for a specific time period for these indicators on a daily basis
#all_results=pd.Dataframe()
all_results=[]
for myid in list1:
    results=fred.get_series(myid,observation_start=start,observation_end=end)
    results =results.to_frame(name=myid)
    all_results.append(results)
```

```python
df5=pd.concat(all_results, axis=1)
```

```python
# bfill values using values from next column
df6=df5.bfill(axis ='columns')
```
