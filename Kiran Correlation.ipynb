```python
import pandas as pd
df = pd.read_csv('Correlation_chart1.csv', header=None)
data = df.values

for i in range(1, 429):
    for j in range(i + 1, 429):
        if (data[i][j] < '0.00' and data[i][j] > '-0.00'):
            print(data[i][0] + "," + data[0][j] + "," + data[i][j])
```

```python
import pandas as pd
df = pd.read_csv('Correlation_chart1.csv', header=None)
data = df.values

for i in range(1, 429):
    for j in range(i + 1, 429):
        if (data[i][j] >= '-0.10' and data[i][j] <= '0.10'): # I want the values in the range of -0.10 ti 0.10 i.e +-10%
            print(data[i][0] + "," + data[0][j] + "," + data[i][j])
```
