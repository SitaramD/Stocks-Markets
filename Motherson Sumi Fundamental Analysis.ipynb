```python
import yfinance as yf
import seaborn as sn
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Downloaded Statements form Screener.in for MotherSumi
df=pd.read_csv("MothersonSumi.csv", index_col =None)
```

```python
df
```

```python
# Dropping Null Values from the dataframe
df=df.dropna()
```

```python
df
```

```python
# Transposing Rows and column
df_t = df.T
```

```python
df_t
```

```python
# set index using column
df_t = df_t.set_index( 1, inplace =False) 
```

```python
df_t 
```

```python
df_t.columns = ['Equity Share Capital','Reserves','Borrowings','Other Liabilities','Total1','Net Block','Capital Work in Progress','Investments','Other Assets','Total','Working Capital','Debtors','Inventory','Debtor Days','Inventory Turnover','Return on Equity']
```

```python
# Naming Index as Year
df_t.index.names = ["Year"]
```

```python
df_t
```

```python
# Dropping first row
df_t = df_t.drop("Narration", axis=0) 
```

```python
df_t
```

```python
# Droppring Total1 column, which is a duplicate column
df_t.drop('Total1', inplace=True, axis=1) 
```

```python
#Code to % remove a character from Return on Equity
df_t["Return on Equity"] = df_t["Return on Equity"]. str. replace(r'\D', '')
```

```python
# Completed Extraction and Cleaning, preprocessing Done
df=df_t
df
```

```python
df.info()
```

```python
# Exploratory Data Analysis for each features
```

```python
# Converting Strings into float variables
df["Equity Share Capital"] = [float(str(i).replace(",", "")) for i in df["Equity Share Capital"]]
df["Reserves"] = [float(str(i).replace(",", "")) for i in df["Reserves"]]
df["Borrowings"] = [float(str(i).replace(",", "")) for i in df["Borrowings"]]
df["Other Liabilities"] = [float(str(i).replace(",", "")) for i in df["Other Liabilities"]]
df["Net Block"] = [float(str(i).replace(",", "")) for i in df["Net Block"]]
df["Capital Work in Progress"] = [float(str(i).replace(",", "")) for i in df["Capital Work in Progress"]]
df["Investments"] = [float(str(i).replace(",", "")) for i in df["Investments"]]
df["Other Assets"] = [float(str(i).replace(",", "")) for i in df["Other Assets"]]
df["Total"] = [float(str(i).replace(",", "")) for i in df["Total"]]
df["Working Capital"] = [float(str(i).replace(",", "")) for i in df["Working Capital"]]
df["Debtors"] = [float(str(i).replace(",", "")) for i in df["Debtors"]]
df["Inventory"] = [float(str(i).replace(",", "")) for i in df["Inventory"]]
df["Debtor Days"] = [float(str(i).replace(",", "")) for i in df["Debtor Days"]]
df["Inventory Turnover"] = [float(str(i).replace(",", "")) for i in df["Total"]]
#df["Return on Equity"] = [float(str(i).replace(",", "")) for i in df["Inventory Turnover"]]
```
