```python
# Importing Libraries
import numpy as np
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the file with 1232 stocks and its corresponding correlation values
df= pd.read_csv("Common Stock NYSE 21012023.csv")
```

```python
df
```

```python
# Converting dataframe to List
Stock_list =df["Symbol"].values.tolist()
```

```python
# Calculation to grab all the closing prices for the  stock list into one DataFrame
closing_df = yf.download(Stock_list, period = "1mo", interval ="1d")['Adj Close']
```

```python
closing_df.tail()
```

```python
df= closing_df.corr().round(2)
df
```

```python
#Now we drop a rows whose all data is missing or contain null values(NaN)
# using dropna() function    
df.dropna(how = 'all', axis=1, inplace =True)
df
```

```python
matrix=df
matrix= matrix.unstack()
k=matrix[matrix == 0.00]
#k.to_csv("NYSE_Cor_1243_17022023.csv")
```

```python
#df2= pd.read_csv("NYSE_Cor_1243_17022023.csv")
df2=k
df2
```

```python
# Counting Duplictaes
dups_count = df2.pivot_table(columns=['Unnamed: 0'], aggfunc='size')
print (dups_count)
```

```python
# Printing duplicates along with count
dups_count
```

```python
# Sorting from high to low, descending order
h= dups_count.sort_values(ascending=False) 
h
```

```python
h.head(15)
```

```python
# Highly incorrelated stocks having maximum zero correlation
column_names = h.keys().values.tolist()
print(column_names[0:15])
```
